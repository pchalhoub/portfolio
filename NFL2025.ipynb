```python
import pandas as pd

df = pd.read_csv('/content/sportsref_download.csv')#.drop(columns=['Unnamed: 45'])
df
```

```python
matchups = df.merge(
    df,
    on=['Week', 'TeamID', 'Opp'],
    suffixes=('', '_OPP')
)
matchups
```

```python
df_train = matchups.copy().dropna()
```

```python
df_train["Rslt"] = df_train["Rslt"].map({"W": 1, "L": 0})
df_train["Rslt_OPP"] = df_train["Rslt_OPP"].map({"W": 1, "L": 0})
```

```python
df_train.columns
```

```python
X = df_train[['TeamID', 'Home/Away', 'Opp', 'Pts', 'PtsO', 'Cmp',
       'Att', 'Cmp%', 'Yds', 'TD', 'Y/A', 'AY/A', 'Rate', 'Sk', 'Yds.1',
       'Att.1', 'Yds.2', 'TD.1', 'Y/A.1', 'Ply', 'Tot', 'Y/P', 'FGA', 'FGM',
       'XPA', 'XPM', 'Pnt', 'Yds.3', 'Pass', 'Rsh', 'Pen', '1stD', '3DConv',
       '3DAtt', '4DConv', '4DAtt', 'Pen.1', 'Yds.4', 'FL', 'Int', 'TO', 'Home/Away_OPP',
       'Pts_OPP', 'PtsO_OPP', 'Cmp_OPP', 'Att_OPP', 'Cmp%_OPP', 'Yds_OPP',
       'TD_OPP', 'Y/A_OPP', 'AY/A_OPP', 'Rate_OPP', 'Sk_OPP', 'Yds.1_OPP',
       'Att.1_OPP', 'Yds.2_OPP', 'TD.1_OPP', 'Y/A.1_OPP', 'Ply_OPP', 'Tot_OPP',
       'Y/P_OPP', 'FGA_OPP', 'FGM_OPP', 'XPA_OPP', 'XPM_OPP', 'Pnt_OPP',
       'Yds.3_OPP', 'Pass_OPP', 'Rsh_OPP', 'Pen_OPP', '1stD_OPP', '3DConv_OPP',
       '3DAtt_OPP', '4DConv_OPP', '4DAtt_OPP', 'Pen.1_OPP', 'Yds.4_OPP',
       'FL_OPP', 'Int_OPP', 'TO_OPP']]
y = df_train[['Rslt']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import log_loss

X_train, X_test, y_train, y_test = train_test_split(X, y.values.ravel(), test_size=0.25, random_state=42)

clf = LogisticRegression(max_iter=2000)
beta = clf.fit(X_train, y_train)

y_pred = beta.predict(X_test)
y_pred_proba = beta.predict_proba(X_test)
loss = log_loss(y_test, y_pred_proba)

print(loss)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.feature_selection import RFE

results = []

for k in range(1, X.shape[1]+1):
    rfe = RFE(estimator=LogisticRegression(max_iter=2000), n_features_to_select=k)
    X_rfe = rfe.fit_transform(X_train, y_train)
    score = cross_val_score(LogisticRegression(max_iter=2000), X_rfe, y_train, cv=5).mean()
    results.append((k, score))

results_df = pd.DataFrame(results, columns=['Num_Features', 'CV_Accuracy'])
print(results_df.sort_values('CV_Accuracy', ascending=False))
```

```python
print(results_df.sort_values('CV_Accuracy', ascending=False).head(30))
```

```python
coef_df = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': clf.coef_[0]
}).sort_values(by='Coefficient', key=abs, ascending=False)

print(coef_df.head(20))
```

```python
X_verify = df_train[['Pts', 'PtsO',
       'FGA', 'FGM',
       'Pass', 'Pen',
       '4DConv', '4DAtt', 'Pen.1', 'TO',
       'Pts_OPP', 'PtsO_OPP', 'FGA_OPP', 'FGM_OPP',
       'Pass_OPP', 'Pen_OPP', '4DConv_OPP', '4DAtt_OPP', 'Pen.1_OPP',
       'TO_OPP']]

X_train, X_test, y_train, y_test = train_test_split(X_verify, y.values.ravel(), test_size=0.25, random_state=42)

clf = LogisticRegression(max_iter=2000)
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
y_pred_proba = clf.predict_proba(X_test)
loss = log_loss(y_test, y_pred_proba)

print(loss)
print("Accuracy:", accuracy_score(y_test, y_pred))
```
