```python
from ultralytics import YOLO
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

model = YOLO('/content/yolo11n-obb.pt')


cap = cv2.VideoCapture(video_path)

frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
line_y_left = 10
line_y_right = frame_width - 10
line_x = frame_height // 2
```

```python
vehicle_count_right_to_left = 0
vehicle_count_left_to_right = 0

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame)

    for result in results[0].obb.xywhr:
      if 10 - 5 <= result[0] <= 10 + 5 and result[1] < line_x - 39:
        vehicle_count_right_to_left += 1
      elif 1270 - 5 <= result[0] <= 1270 + 5 and result[1] > line_x - 39:
        vehicle_count_left_to_right += 1

    cv2.line(frame, (10, 0), (10, frame_height), (0, 255, 255), 2)
    cv2.line(frame, (1270, 0), (1270, frame_height), (0, 255, 255), 2)
    cv2.line(frame, (0, line_x - 39), (frame_width, line_x - 39), (0, 255, 255), 2)
    cv2.putText(frame, f"Vehicle Count Right-Left: {vehicle_count_right_to_left}", (10, 50),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    cv2.putText(frame, f"Vehicle Count Left-Right: {vehicle_count_left_to_right}", (10, 150),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    #cv2_imshow(frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()

print("Total vehicle count right-left:", vehicle_count_right_to_left)
print("Total vehicle count left-right:", vehicle_count_left_to_right)
```
